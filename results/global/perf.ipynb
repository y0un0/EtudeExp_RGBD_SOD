```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('result.txt', sep=',')
df = df.round(3)
df
```

```python
cat = df[df['Dataset'] == 'Complexbackground'].reset_index()
cat
```

```python
plt.scatter(cat['S-measure_alpha05'][0], cat['max-fmeasure'][0], label='DANet', s=104)
plt.scatter(cat['S-measure_alpha05'][1], cat['max-fmeasure'][1], label='DFMNet', s=9)
plt.scatter(cat['S-measure_alpha05'][2], cat['max-fmeasure'][2], label='JL-DCF', s=561)
plt.scatter(cat['S-measure_alpha05'][3], cat['max-fmeasure'][3], label='UCNet', s=123)
# plt.scatter(cat['S-measure_alpha05'][0], cat['max-fmeasure'][0], label='DANet')
# plt.scatter(cat['S-measure_alpha05'][1], cat['max-fmeasure'][1], label='DFMNet')
# plt.scatter(cat['S-measure_alpha05'][2], cat['max-fmeasure'][2], label='JL-DCF')
# plt.scatter(cat['S-measure_alpha05'][3], cat['max-fmeasure'][3], label='UCNet')
plt.annotate("DANet", (cat['S-measure_alpha05'][0], cat['max-fmeasure'][0]))
plt.annotate("DFMNet", (cat['S-measure_alpha05'][1], cat['max-fmeasure'][1]))
plt.annotate("JL-DCF", (cat['S-measure_alpha05'][2], cat['max-fmeasure'][2]))
plt.annotate("UCNet", (cat['S-measure_alpha05'][3], cat['max-fmeasure'][3]))
plt.ylabel("Accuracy (maxF-measure)")
plt.xlabel("Structure similarity")
plt.title("Complexbackground")
plt.grid()
plt.show()
```
